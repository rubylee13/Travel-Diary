```python
import numpy as np
import datetime
import dateutil.parser

def getTimeofDayDistribution(purpose):
    
    purposeSegments = list(segments.find({'mainPurpose': purpose}))
    frequency = np.zeros((24))
    
    for segment in purposeSegments:
        startTime = dateutil.parser.parse(segment['startTime'])
        endTime = dateutil.parser.parse(segment['endTime'])
        while startTime < endTime:
            frequency[startTime.hour] += 1
            startTime = startTime + datetime.timedelta(0, 0, 0, 0, 0, 1, 0)
    return frequency
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nPurposes = len(purposes)
counter = 0
plt.figure(num = 1, figsize = (16, 5))  
for purpose in purposes: 
    frequency = getTimeofDayDistribution(purpose)
    plt.subplot(1, nPurposes, counter)        
    plt.bar(np.arange(0, 24, 1), frequency, color = 'k')
    plt.xlim(0, 24)
    #plt.ylim(0, 60)
    plt.xticks(np.arange(0, 25, 6), fontsize = 10)
    plt.yticks(fontsize = 10)
    plt.title(purpose, fontsize = 10)
    plt.xlabel('Time of day', fontsize = 10) 
    counter += 1
```

```python
def getDayOfWeekDistribution(purpose):
    
    purposeSegments = list(segments.find({'mainPurpose': purpose}))
    frequency = np.zeros((7))
    weekDays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    
    for segment in purposeSegments:
        startTime = dateutil.parser.parse(segment['startTime'])
        endTime = dateutil.parser.parse(segment['endTime'])
        while startTime < endTime:
            currentDay = startTime.strftime('%A')
            frequency[weekDays.index(currentDay)] += 1
            startTime = startTime + datetime.timedelta(1)
    return frequency
```

```python
nPurposes = len(purposes)
counter = 0
plt.figure(num = 1, figsize = (16, 5))  
for purpose in purposes: 
    frequency = getDayOfWeekDistribution(purpose)
    plt.subplot(1, nPurposes, counter)        
    plt.bar(np.arange(0, 7, 1), frequency, color = 'k')
    #plt.xlim(0, 24)
    #plt.ylim(0, 60)
    #plt.xticks(np.arange(0, 25, 6), fontsize = 10)
    plt.yticks(fontsize = 10)
    plt.title(purpose, fontsize = 10)
    plt.xlabel('Day of week', fontsize = 10) 
    counter += 1
```

```python
duration, counter = {}, 0
plt.figure(num = 1, figsize = (16, 5))  

for purpose in purposes:
    
    duration[purpose] = []
    query = {'mainPurpose': purpose}
    projection = {'_id': 0, 'duration': 1}
    purposeDurations = list(segments.find(query, projection))
    for purposeDuration in purposeDurations:
        duration[purpose].append(purposeDuration['duration'])
    
    plt.subplot(1, nPurposes, counter)
    tPurpose = np.zeros(len(duration[purpose]))
    for i in range(0, len(duration[purpose])):
        tPurpose[i] = duration[purpose][i]/3600
    if len(duration[purpose]) > 0:
        plt.hist(tPurpose, normed = 1, color = 'k')
    plt.yticks(fontsize = 10)
    plt.title(purpose, fontsize = 10)
    plt.xlabel('Number of hours', fontsize = 10) 
    counter += 1
```

```python
def getDurationDistribution(phNum):
    
    activities = list(segments.find({'type': 'Activity', 'phNum': phNum}))
    nearbyDurations, totalDuration = [], 0
    
    for activity in activities:
        totalDuration += activity['duration']
        nearbyDuration = 0
        centroid = activity['centroid']['coordinates']
        query = {'centroid': {'$nearSphere': {'$geometry': {'type': 'Point',
                                                            'coordinates': [centroid[0], centroid[1]]},
                                              '$maxDistance': 100}},
                 'phNum': phNum}
        nearbyActivities = segments.find(query)
        for nearbyActivity in nearbyActivities:
            nearbyDuration += nearbyActivity['duration']
        nearbyDurations.append(nearbyDuration)
    
    for i in range(0, len(nearbyDurations)):
        nearbyDurations[i] = (100 * nearbyDurations[i]) / totalDuration
        
    duration = {}
    for purpose in purposes:
        duration[purpose] = []

    for i in range(0, len(activities)):
        if 'mainPurpose' in activities[i]:
            duration[activities[i]['mainPurpose']].append(nearbyDurations[i])
    
    return duration
```

```python
def plotDurationHistogram(duration):
    nPurposes = len(purposes)
    counter = 0
    plt.figure(num = 1, figsize = (16, 5))  
    for purpose in purposes: 
        plt.subplot(1, nPurposes, counter)
        tPurpose = np.zeros(len(duration[purpose]))
        for i in range(0, len(duration[purpose])):
            tPurpose[i] = duration[purpose][i]
        if len(duration[purpose]) > 0:
            plt.hist(tPurpose, normed = 1, color = 'k')
        plt.xlim(0, 100)
        #plt.ylim(0, 60)
        plt.xticks(np.arange(0, 101, 25), fontsize = 10)
        plt.yticks(fontsize = 10)
        plt.title(purpose, fontsize = 10)
        plt.xlabel('%age of total time', fontsize = 10) 
        counter += 1
```

```python
durationTotal = {}
for purpose in purposes:
    durationTotal[purpose] = []

phNums = segments.distinct('phNum')
for phNum in phNums:
    durationPhNum = getDurationDistribution(phNum)
    for purpose in purposes:
        durationTotal[purpose] += durationPhNum[purpose]
```

```python
plotDurationHistogram(durationTotal)
```

# Feature Construction

```python
def getTimeofDay(segment):
    
    frequency = np.zeros((24))    
    startTime = dateutil.parser.parse(segment['startTime'])
    endTime = dateutil.parser.parse(segment['endTime'])
    while startTime < endTime:
        frequency[startTime.hour] += 1
        startTime = startTime + datetime.timedelta(0, 0, 0, 0, 0, 1, 0)    
    return frequency
```

```python
def getDayOfWeek(segment):

    frequency = np.zeros((7))    
    weekDays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']    
    startTime = dateutil.parser.parse(segment['startTime'])
    endTime = dateutil.parser.parse(segment['endTime'])
    while startTime < endTime:
        currentDay = startTime.strftime('%A')
        frequency[weekDays.index(currentDay)] += 1
        startTime = startTime + datetime.timedelta(1)
    return frequency
```

```python
def getDurationInNeighborhood(activity):
    
    nearbyDuration = 0
    centroid = activity['centroid']['coordinates']
    query = {'centroid': {'$nearSphere': {'$geometry': {'type': 'Point',
                                                        'coordinates': [centroid[0], centroid[1]]},
                                          '$maxDistance': 100}},
             'phNum': activity['phNum']}
    nearbyActivities = segments.find(query)
    for nearbyActivity in nearbyActivities:
        nearbyDuration += nearbyActivity['duration']    
    return nearbyDuration
```

```python
def getTotalDuration(activities):
    
    duration = 0
    for activity in activities:
        duration += activity['duration']
    return duration        
```

```python
def calFeatures(segments, phNums):
    
    numActivities = segments.find({'type': 'Activity', 'mainPurpose': {'$exists': True}}).count()
    features = np.zeros((numActivities, 34))
    count = 0
    purposes = segments.distinct('mainPurpose')

    for phNum in phNums:
        query = {'type': 'Activity', 'phNum': phNum, 'mainPurpose': {'$exists': True}}
        projection = {'_id': 0, 'trackPoints': 0}
        activities = list(segments.find(query, projection))
        totalDuration = getTotalDuration(activities)
        for activity in activities:            
             
            timeOfDay = getTimeofDay(activity)
            dayOfWeek = getDayOfWeek(activity)
            durationActivity = activity['duration']
            durationInNeighborhood = (100 * getDurationInNeighborhood(activity)) / totalDuration

            features[count, :timeOfDay.shape[0]] = timeOfDay
            features[count, timeOfDay.shape[0]:timeOfDay.shape[0] + dayOfWeek.shape[0]] = dayOfWeek
            features[count, -3] = durationActivity
            features[count, -2] = durationInNeighborhood             
            features[count, -1] = purposes.index(activity['mainPurpose']) + 1
            count += 1
    
    return features
```

```python
featureLabels = ['12-1 AM', '1-2 AM', '2-3 AM', '3-4 AM', '4-5 AM', '5-6 AM', '6-7 AM', '7-8 AM', '8-9 AM', '9-10 AM',
                 '10-11 AM', '11-12 PM', '12-1 PM', '1-2 PM', '2-3 PM', '3-4 PM', '4-5 PM', '5-6 PM', '6-7 PM', '7-8 PM',
                 '8-9 PM', '9-10 PM', '10-11 PM', '11-12 AM', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',
                 'Saturday', 'Sunday', 'Length of activity', 'Time spent in neighborhood', 'Activity Purpose']
features = calFeatures(segments, phNums)
```

# Dimension Reduction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 1, whiten = True)
pca.fit(features[:, :-1])
print(pca.explained_variance_ratio_) 
x = pca.transform(features[:, :-1]).flatten()
y = features[:, -1]
print x.shape, y.shape
plt.scatter(x, y)
```

# Inference

```python
from sklearn import tree

featuresEstimation = features[:int(0.9 * np.size(features, 0)), :]
featuresValidation = features[int(0.9 * np.size(features, 0)):, :]

clf = tree.DecisionTreeClassifier(max_depth = 10, min_samples_leaf = 5)
clf = clf.fit(featuresEstimation[:, :-1], featuresEstimation[:, -1])
infAccuracy = (100.0 * sum(clf.predict(featuresValidation[:, :-1]) == featuresValidation[:, -1])) / np.size(featuresValidation, 0)
print 'Inference accuracy: %.2f%%' %infAccuracy
```

```python
from sklearn import ensemble
from sklearn import cross_validation

forestClf = ensemble.RandomForestClassifier()
forestScores = cross_validation.cross_val_score(forestClf, features[:, :-1], features[:, -1], cv = 5)

for (i, forestScore) in enumerate(forestScores.tolist()):
    print 'Inference accuracy for %s: %.1f%%' %(purposes[i], 100 * forestScore)
print '\nTotal Inference accuracy: %.1f%%\n' %(100 * forestScores.mean())

forestClf.fit(features[:, :-1], features[:, -1])
for (i, importance) in enumerate(forestClf.feature_importances_):
    print '%s: %.1f%%' %(featureLabels[i], 100 * importance)
```
